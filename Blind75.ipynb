```python
#TwoSum
# Input: nums = [2,7,11,15], target = 9
# Output: [0,1]
# Output: Because nums[0] + nums[1] == 9, we return [0, 1].

nums = [2,7,11,15]
target = 9
def TwoSum(nums,target):
    #HashMap (Dictionary)
    temp={}
    #Enumerate index and number
    for i,n in enumerate(nums):
        #Complement
        complement=target-n
        #Check if it is present in dictionary
        if complement in temp:
            #If it is return the indexes of the complement(from the dictionary) and the current number
            return [temp[complement],i]
        #Store the current number(key) and its index(value) in the dictonary
        temp[n]=i
TwoSum(nums,target)
```

```python
#Three Sum
# Input: nums = [-1,0,1,2,-1,-4]
# Output: [[-1,-1,2],[-1,0,1]]


# Given an integer array nums, 
# return all the triplets [nums[i], nums[j], nums[k]] 
# such that i != j, i != k, and j != k, 
# and nums[i] + nums[j] + nums[k] == 0.

# Notice that the solution set must not contain duplicate triplets.

 
nums = [34,55,79,28,46,33,2,48,31,-3,84,71,52,-3,93,15,21,-43,57,-6,86,56,94,74,83,-14,28,-66,46,-49,62,-11,43,65,77,12,47,61,26,1,13,29,55,-82,76,26,15,-29,36,-29,10,-70,69,17,49]

def threeSum(nums):
    #Sort yooo << Required to eliminate duplicate cases
    nums.sort()
    res = []
    #One Fixed Pointer
    for i in range(len(nums)):
        #Eliminate duplicate cases
        #Dont want cases for same i since we must have already found them in a previous iteration 
        if(i>0 and nums[i]==nums[i-1]):
            continue
        #Two Pointers
        left = i+1
        right = len(nums) - 1
        
        while(left<right):
            #Sum i+j+k
            s = nums[i] + nums[left] + nums[right]
            #Since array is sorted
            #Increase left if less than zero to reach zero
            if(s<0):
                left +=1
            #Decrease right if greater than zero to reach zero    
            elif(s>0): 
                right-=1
            #Sum is now zero
            else:

                #Append i,j,k to result
                res.append([nums[i],nums[left],nums[right]])


                #Since we found our answer with fixed i and we dont another answer with same j,k
                #Eliminiate duplicate cases
                while(left<right and nums[left]==nums[left+1]):
                    left +=1
                while(left<right and nums[right]==nums[right-1]):
                    right-=1

                #Incrementing once since its still on duplicate case due to our condition
                left+=1
                right-=1

    return res

threeSum(nums)
```

```python
# Best Time to Buy and Sell Stock III
# Max Profit With atmost 2 Transactions K=2

# Input: prices = [3,3,5,0,0,3,1,4]
# Output: 6
# Explanation: Buy on day 4 (price = 0) and sell on day 6 (price = 3), profit = 3-0 = 3.
# Then buy on day 7 (price = 1) and sell on day 8 (price = 4), profit = 4-1 = 3.

#Divide and Conquer for two transactions

def maxProfit(prices):
    n = len(prices)
    #Edge Case
    if (n<2):
        return 0
    
    #Calculate the max profit going from the left
    left = [0]*n
    leftMin = prices[0]
    for i in range(1,n):
        #Store the max profit uptil now
        left[i]=max(left[i-1],prices[i]-leftMin)
        #Update the min
        leftMin=min(leftMin,prices[i])
        
    #Calculate the max profit coming back from the right
    right = [0]*n
    rightMax = prices[n-1]
    for i in range(n-2,-1,-1):
        #Store the max profit uptil now
        right[i]=max(right[i+1],rightMax-prices[i])
        #Update the max
        rightMax=max(rightMax,prices[i])
    
    #Divide the array, store the profit which gives maximum profit
    profit=right[0]
    for i in range(1,n):
        profit = max(profit,left[i-1]+right[i])
        
    return profit
```

```python
# Best Time to Buy and Sell Stock IV
# Max Profit With atmost K Transactions

# Input: k = 2, prices = [2,4,1]
# Output: 2
# Explanation: Buy on day 1 (price = 2) and sell on day 2 (price = 4), profit = 4-2 = 2.

#Dynamic Programming
# Formula 
#d = day in prices and t is the transaction(k)
# profit[t][d] = max(profit[t-1][d-1], prices[d] + max(profit[t-1][x-1] - prices[x-1])) [WHERE 0<=x<d]
# Either you sell the stock or you dont sell( in this case the profit remains the same)


# profit[t-1][d-1] >>>> You dont sell the stock, so the profit is same as the previous one


# prices[d] + max(profit[t-1][x-1] - prices[x-1]) >>>> You sell the stock, so we add the previous profit
                                                        #and the price at which we buy again.
    
# prices[d] >>>> The price at which we sell
# max(profit[t-1][x-1] - prices[x-1]) >> The profit we had before buying - the price at which we had bought before 


  
k = 2
prices = [2,4,1]
def maxProfit(k, prices):
    
    #Edge Case
    if not len(prices):
        return 0
    
    
    #Intialization 
    #0 Profit when 0 transactions t=0 
    #0 Profit on 1st day since nothing is sold d=0
    profit = [[0 for d in prices] for t in range(k+1)]
    
    
    for t in range(1,k+1):
        
        #For finding max, initializing to -INF
        maxProfitUntilNow = float("-inf")
        
        for d in range(1,len(prices)):
            
            #Max Profit Until Now 
            #We calcualte this for each step and store the point at which we get max profit
            maxProfitUntilNow = max(maxProfitUntilNow , profit[t-1][d-1] - prices[d-1]) #Buy karke rakha he
              
            
            #Decision >> Whether we dont sell, or sell
            profit[t][d] = max(profit[t][d-1] , prices[d] + maxProfitUntilNow) #Bus sell karke dekhna kaunsa max he
            
            
    #Return the last place in the matrix
    return profit[-1][-1]


maxProfit(k, prices)
```

```python
# Product of Array Except Self
#https://leetcode.com/problems/product-of-array-except-self/discuss/1342916/3-Minute-Read-Mimicking-an-Interview
# Input: nums = [1,2,3,4]
# Output: [24,12,8,6]
    
nums = [1,2,3,4]    

#Prefix Sum Method
def productExceptSelf(nums):
    n = len(nums)
    temp = [1]*n
    #Product of all from numbers the left side
    for i in range(1,n):
        temp[i] =  temp[i-1]*nums[i-1]
                                                    
    #Product of all numbers from the right side
    r=1
    for i in range(n-1,-1,-1):
        temp[i] =  temp[i] * r
        r = r * nums[i]

    return temp

productExceptSelf(nums)
```

```python
# Maximum Subarray

# Input: nums = [-2,1,-3,4,-1,2,1,-5,4]
# Output: 6
# Explanation: [4,-1,2,1] has the largest sum = 6.
    
nums = [-2,1,-3,4,-1,2,1,-5,4] 
#Brute
def maxSubArrayBrute(nums):
    globalSum = float('-inf')
    for i in range(len(nums)):
        localSum = 0
        for j in range(i,len(nums)):
            localSum += nums[j]
            if(localSum > globalSum):
                globalSum = localSum
    return globalSum
            
#Kadanes 
def maxSubArray(nums):
    localSum = float('-inf')
    globalSum = localSum
    for i in range(len(nums)): 
        #Do we start over OR Add to the current Sum?
        #That is we nums[i] is itself greater than the addition? 
        #If so than start over
        #If not then go on adding
        localSum = max(nums[i], localSum + nums[i])
        globalSum = max(localSum,globalSum)
    return globalSum

print(maxSubArray(nums) , maxSubArrayBrute(nums)) 
```

```python
# Maximum Product Subarray

# Input: nums = [2,3,-2,4]
# Output: 6
# Explanation: [2,3] has the largest product 6.

nums = [2,3,-2,4]


#Keep track of localMax and localMin 
def maxProduct(nums):
    #Initializing to 1 since its a product
    
    localMin = 1
    localMax = 1
    globalMax = float('-inf')
    
    for n in nums:
        #Since localMax gets updated, we need the old localMax
        temp = localMax
        
        #This is done to handle the negative numbers since multiplying two negative numbers results in a positive number
        #Cant directly use Kadane's because of this
        #So we keep track of the localMax and the localMin
        #(Local Min could turn into positive if there's a negative number ahead and can be greater than the localMax)
        
        #Either max of  n * localMax OR n * localMin OR we reset the subarray
        localMax = max(n * localMax, n * localMin, n)
        #Use the old localMax stored in temp since it gets updated
        #Same explanation as above
        localMin = min(n * temp, n * localMin, n)
        
        globalMax = max(globalMax, localMax, localMin)
    return globalMax
```

```python
# Find Minimum in Rotated Sorted Array

# Input: nums = [3,4,5,1,2]
# Output: 1
# Explanation: The original array was [1,2,3,4,5] rotated 3 times.
    
nums = [3,4,5,1,2]

#Binary Search
def findMin(nums):
    def BinarySearch(nums,low,high):
        if(low<=high):
            #Means the array is already sorted correctly and the first element is the minimum
            if(nums[low]<=nums[high]):
                return nums[low]
            else:
                mid = (low + high)//2
                #Min element lies on the side which is not sorted
                #Can be found by comparing mid with low and high whether which side is sorted
                #Right side not sorted
                if (nums[mid] >= nums[low] and nums[mid] >= nums[high]):
                    return BinarySearch(nums,mid+1,high)
                #Left side not sorted
                elif (nums[mid] <= nums[low] and nums[mid] <= nums[high]):
                    return BinarySearch(nums,low,mid)

    return BinarySearch(nums,0,len(nums)-1)

def findMinIterative(nums):
    left, right = 0, len(nums) - 1
    while nums[left] > nums[right]:
        middle  = (left + right) // 2
        if nums[middle] < nums[right]:
            right = middle
        else:
            left = middle + 1
    return nums[left]
            

findMin(nums)
```

```python
# Search in Rotated Sorted Array

# Input: nums = [4,5,6,7,0,1,2], target = 0
# Output: 4

nums = [4,5,6,7,0,1,2]
target = 0

#Binary Search
def search(nums,target):
    left = 0
    right = len(nums)-1
    while(right>=left):
        mid = left + (right-left)//2
        #Base Condition
        if(nums[mid]==target):
            return mid
        #We would have 4 cases, due to the rotated structure of the array depending on the target and the middle element.
        #Just simulate those cases and write conditions accordingly.
        else:
            #If sorted section is on the left
            if(nums[left]<=nums[mid]):
                #Target greater than mid but less than left
                if(target > nums[mid] or target < nums[left]):
                    left = mid + 1
                #Target between left and mid
                else:
                    right = mid - 1
            #If sorted section is on the left
            else:
                #Target less than mid but greater than right
                if(target < nums[mid] or target > nums[right]):
                    right = mid - 1 
                #Target between mid and right
                else:
                    left = mid + 1

    return -1
```

```python
# Container With Most Water

# Input: height = [1,8,6,2,5,4,8,3,7]
# Output: 49

height = [1,8,6,2,5,4,8,3,7]

#Two Pointers
def maxArea(height):
    n=len(height)
    i=0
    j=n-1
    maxWater=0
    while(i<j):
        #Basic Greedy             
        #Area Formula
        #Difference between the indexes and the minimum of two numbers. 
        #Since its an area and cant have the bigger one. (WATER OVERFLOW)
        maxWater = max(maxWater, (j-i) * min(height[i],height[j]))
        #Adjust pointers if they are low
        if(height[i]<height[j]):
            i+=1
        else:
            j-=1
    return maxWater

maxArea(height)

```

```python
# 424. Longest Repeating Character Replacement

# Input: s = "ABAB", k = 2
# Output: 4
# Explanation: Replace the two 'A's with two 'B's or vice versa.

#Sliding Window
#Condition for window is that end - start + 1 - FrequencyOfMaxCharInCurrentWindow <=k
#Only then we can substitue the non max frequency characters in the substring
def characterReplacement(s,k):
    i=0
    j=0
    temp = {}
    maxLen=0
    while(j < len(s)):
        #Adding the count to dictionary
        if s[j] not in temp:
            temp[s[j]]=0
        temp[s[j]]+=1
        
        #Count of the maximum frqeuncy character in the substring
        #Increment i until the window condition is satisfied
        while(j - i + 1 - max(temp.values()) > k):
            #Decrement the count in dictionary
            temp[s[i]]-=1
            i+=1
        
        #MaxLen of the substring 
        maxLen=max(maxLen, j - i + 1)
        j+=1
    return maxLen

characterReplacement('ABAB',2)
```

```python
# 76. Minimum Window Substring

# Input: s = "ADOBECODEBANC", t = "ABC"
# Output: "BANC"
# Explanation: The minimum window substring "BANC" includes 'A', 'B', and 'C' from string t.
    
s = "ADOBECODEBANC"
t = "ABC"


#Sliding Window
def minWindow(s,t):
    
    tDict = {}
    sDict = {}
    
    #Dictionary for string t. Will remain static.
    for c in t:
        if c not in tDict:
            tDict[c]=0
        tDict[c]+=1
        
    
    resLen = float('inf') #Track min. len. of the substring
    start,end = -1,-1 #Track start and end index of the substring
    i,j = 0,0 #Two pointers for the window
    
    #Have is the characters which required that we have in the current window
    #Need is the characters that are needed in the window in the substring
    #This is required to establish that the necessary conditions that we have formed out susbtring is met
    have,need = 0,len(tDict) #len(tDict) that means only unique characters
    
    
    while(j<len(s)):
        
        #Dictionary for string s. Will change the count as the window changes.
        if s[j] not in sDict:
            sDict[s[j]]=0
        sDict[s[j]]+=1

        #If the characters is in t and also we have the required frequency.
        if s[j] in tDict and sDict[s[j]]==tDict[s[j]]:
            have+=1
        
        #If we have the required characters.
        while have == need:
            
            #Update the min. length and the start,end indices 
            if(resLen > j - i + 1):
                resLen = j - i + 1
                start = i
                end = j
             
            #Decrease window length
            #Now we start popping each character from the left and see if have any uneccessary characters to get the min. substring     
            sDict[s[i]]-=1
            #If it is a character in t, then changing the window removes it from the substring or doesnt meet the required count
            if s[i] in tDict and sDict[s[i]] < tDict[s[i]]:
                have-=1
            #Increment left
            i+=1
        #Increment right     
        j+=1
    return s[start:end+1]


minWindow(s,t)    
```

```python
# 125. Valid Palindrome

# Input: s = "A man, a plan, a canal: Panama"
# Output: true
# Explanation: "amanaplanacanalpanama" is a palindrome.

def isPalindrome(s):
    # temp =""
    # for i in s:
    #     if i.isalnum():
    #         temp+=i.lower()
    # return temp==temp[::-1]

    l, r = 0, len(s)-1
    while l < r:
        
        #Adjust until alphanum
        while l < r and not s[l].isalnum():
            l += 1
        while l <r and not s[r].isalnum():
            r -= 1
        #If not equal return false
        if s[l].lower() != s[r].lower():
            return False
        l +=1; r -= 1
    return True
```

```python
# 5. Longest Palindromic Substring

# Input: s = "babad"
# Output: "bab"
# Explanation: "aba" is also a valid answer.

s = "babad"
def longestPalindrome(s):
    #Expand from middle
    def expandFromMiddle(i,j):
        if(i>j):
            return 0
        #From middle, decrement i and increment j if the characters are equal
        while(i>=0 and j < len(s) and s[i] == s[j]):
            i -= 1
            j += 1
        #Exits if it stops being palindrome
        #Adjusting i and j, since it got incremented/decremented in the last iteration since s[i]!=s[j]
        return (j-1)-(i+1)+1
    
    #Pointers to keep track
    left = 0
    right = 0
    for i in range(len(s)):
        
        #Even Case
        l1 = expandFromMiddle(i,i+1)
        #Odd Case
        l2 = expandFromMiddle(i,i)
        #Max of two cases
        l = max(l1,l2)
        
        #Updating the max
        if(l > right - left):

            left = i - (l-1)//2 #This is important for even case
            right = i + l//2 #Same for both odd and even

    return s[left:right+1]

longestPalindrome(s)
```

```python
# 518. Coin Change 2

# Input: amount = 5, coins = [1,2,5]
# Output: 4
# Explanation: there are four ways to make up the amount:
# 5=5
# 5=2+2+1
# 5=2+1+1+1
# 5=1+1+1+1+1

#2D
def change(amount,coins):
    n = len(coins)
    #Base
    dp = [[0 for x in range(amount+1)] for x in range(n+1)]
    
    #Base : 1 way when amount = 0
    for i in range(1,n+1):
        dp[i][0] = 1

    for i in range(1,n+1):
        for j in range(1,amount+1):
            if coins[i-1] <= j:
                #Addition of both the cases gives us the number of ways
                dp[i][j] = dp[i][j-coins[i-1]] + dp[i-1][j]
            else:
                dp[i][j] = dp[i-1][j]

    return dp[-1][-1]

#1D
def change(amount,coins):
    dp = [1] + [0] * amount
    
    #For each coin
    for coin in coins:
        #We calculate for the respective targets using that many coins
        for i in range(1,amount+1):
            
            #Choice : Take only if less i.e dp[i] + dp[i-coin]
            #If not just dp[i]
            if(coin <= i):

                dp[i] += dp[i-coin]

    return dp[-1]
```

```python
# 57. Insert Interval

# Input: intervals = [[1,3],[6,9]], newInterval = [2,5]
# Output: [[1,5],[6,9]]
    
intervals = [[1,3],[6,9]]
newInterval = [2,5]
    
def insert(intervals, newInterval):

    res = []

    for i in range(len(intervals)):
        #Append is newEnd < Start (Doesnt Overlap!)
        if newInterval[1] < intervals[i][0]:
            #Append since cant overlap with anything else as the array is sorted
            res.append(newInterval)
            #Return as we are down inserting
            return res + intervals[i:]
        #newStart > End (Doesnt Overlap!)
        elif newInterval[0] > intervals[i][1]:
            #Append the current interval
            #New interval could still overlap with future intervals
            res.append(intervals[i])
        #Overlaps
        else:
            #Min of newStart, Start
            newInterval[0] = min(newInterval[0],intervals[i][0])
            #Max of newEnd, End
            newInterval[1] = max(newInterval[1],intervals[i][1])

    res.append(newInterval)
    return res
```

```python
# 253. Meeting Rooms II

# Input: intervals = [[0,30],[5,10],[15,20]]
# Output: 2

# intervals = [[0,30],[5,10],[15,20]]
intervals = [[7,10],[2,4]]

import heapq
def minMeetingRooms(intervals):
    
    #Sort as per start time
    intervals.sort(key = lambda x : x[0])
    
    #Min PQ
    heap = []
  
    for i in range(0,len(intervals)):
          
        #If possible to add in already exisiting classroom with min end time
        if  heap and heap[0] < intervals[i][0]:
            #Replace the max end time of that classroom
            heapq.heapreplace(heap, intervals[i][1])
            
        else:
            #Create a  new class room > Push the end time
            heapq.heappush(heap, intervals[i][1])
                
            
    #No of classrooms
    return len(heap)

minMeetingRooms(intervals)
```

```python
# 729. My Calendar I

# Input
# ["MyCalendar", "book", "book", "book"]
# [[], [10, 20], [15, 25], [20, 30]]
# Output
# [null, true, false, true]

#Double booking not allowed

class MyCalendar:

    def __init__(self):
        self.bookings = []

    def book(self, start: int, end: int) -> bool:
        
        for s,e in self.bookings: 
            #If overlapping
            if max(s,start) < min(e,end):
                return False
        #Add if not, add
        self.bookings.append([start,end])
        return True
    
    
import bisect
intervals = []

def book(start, end):
    
    if end <= start:
        return False
    
    #Right most position to insert
    i = bisect.bisect_right(intervals, start)
      
    #Should be even cause start,end 
    if i % 2!=0:            
        return False
    
    #Left most position to insert
    j = bisect.bisect_left(intervals, end)
    
    #Should be same, if not it already exists for that time
    if i != j:
        return False
    
    #Code for inserting directly via slicing
    intervals[i:i] = [start, end]
    return True


res = []
res.append(book(10,20))
res.append(book(15,25))
res.append(book(30,40))
print(res)


# 731. My Calendar II

# Input
# ["MyCalendarTwo", "book", "book", "book", "book", "book", "book"]
# [[], [10, 20], [50, 60], [10, 40], [5, 15], [5, 10], [25, 55]]
# Output
# [null, true, true, true, false, true, true]

#Triple booking not allowed

class MyCalendarTwo:

    def __init__(self):
        #Bookings
        self.bookings = []
        #Double Bookings
        self.overlaps = []
        
    def book(self, start: int, end: int) -> bool:
        
        #Check if its double booked
        for s,e in self.overlaps:
            if max(start,s) < min(end,e):
                return False
        #Check if its booked    
        for s,e in self.bookings:
            if max(start,s) < min(end,e):
                #If booked, add to double booked array the exact overlap between two
                self.overlaps.append([max(start,s),min(end,e)])
        
        #No booking
        self.bookings.append([start,end])
        return True
```

```python
#Problems on Linked List

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next



# 206. Reverse Linked List
# Input: head = [1,2,3,4,5]
# Output: [5,4,3,2,1]

#Simple manipulation
def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
    #None because last nodes next is None
    prev = None
    while head:
        curr = head
        #Going forward
        head = head.next
        #Attaching to the previous
        curr.next = prev
        #Changing previous
        prev = curr
    return prev

# 141. Linked List Cycle
# Input: head = [3,2,0,-4], pos = 1
# Output: true
# Explanation: There is a cycle in the linked list, where the tail connects to the 1st node (0-indexed).

#Slow and Fast pointers
#Hare and Tortise

#If the fast pointers encounter the slow pointer again, then there is obviously a cycle present.

def hasCycle(self, head: Optional[ListNode]) -> bool:
    
    #Empty List
    if head is None:
        return False
    
    #Slow
    slow = head
    #Fast
    fast = head.next
    
    #Loop until they are not equal
    while slow is not fast :
        
        #If fast reaches end then there is no cycle
        if fast is None or fast.next is None:
            return False
        #Slow
        slow = slow.next
        #Fast
        fast = fast.next.next
    return True


# 21. Merge Two Sorted Lists
# Input: list1 = [1,2,4], list2 = [1,3,4]
# Output: [1,1,2,3,4,4]

#We dont have to preserve the original lists
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        
        #Start temporary node for storing, since we need next
        temp = current = ListNode()
        
        
        while list1 and list2:
            
            #Append whichever is less
            if list1.val < list2.val:
                current.next = list1
                list1 = list1.next
            else:
                current.next = list2
                list2=list2.next
            
            current = current.next
            
        #Whichever list has nodes left, just join directly
        current.next = list1 or list2            
        return temp.next   
    
# 23. Merge k Sorted Lists

# Input: lists = [[1,4,5],[1,3,4],[2,6]]
# Output: [1,1,2,3,4,4,5,6]
# Explanation: The linked-lists are:
# [
#   1->4->5,
#   1->3->4,
#   2->6
# ]
# merging them into one sorted list:
# 1->1->2->3->4->4->5->6

#PQ Method
#This works because the lists are already sorted
def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
    
    #Dummy node for storage
    dummy = curr = ListNode()
    
    #PQ for lowest node of each list
    pq = PriorityQueue()
    
    
    for i,node in enumerate(lists):
        #Put the lowest node in (val,i, node)
        #i needed to break ties as python uses the next value to break ties
        #Since node objects cannot compared, unless you have lt implemented in the class
        if node:
            pq.put((node.val,i,node))
            
    #While PQ is not empty
    while pq.qsize() > 0:
        #Pop from start of the queue the min value
        temp = pq.get()
        #The node
        curr.next = temp[2]
        #The index for storing the next value
        i = temp[1]
        curr = curr.next
        #Store the next value in the same list in the PQ
        if curr.next:
            pq.put((curr.next.val,i,curr.next))
    
    return dummy.next 


# 19. Remove Nth Node From End of List

# Input: head = [1,2,3,4,5], n = 2
# Output: [1,2,3,5]

#2 Positions starting from the end of the list
#Use fast and slow pointer approach to locate the point
def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
    fast = slow = head
    
    #Space fast pointer from slow pointer by n spaces
    for x in range(n):
        fast = fast.next

    #If end is reached by trying to space, then we have to delete to the starting node
    if fast is None:
        return head.next

    #Now, move both the pointers until fast is at the last space, so slow is one before the one we have to delete
    while fast.next:
        fast = fast.next
        slow = slow.next
    
    #Delete the next of slow
    slow.next = slow.next.next
    return head


# 143. Reorder List

# You are given the head of a singly linked-list. The list can be represented as:

# L0 → L1 → … → Ln - 1 → Ln
# Reorder the list to be on the following form:

# L0 → Ln → L1 → Ln - 1 → L2 → Ln - 2 → …
# You may not modify the values in the list's nodes. Only nodes themselves may be changed.


# Input: head = [1,2,3,4]
# Output: [1,4,2,3]

#1. Locate the mid point. Cutting of the list in two parts
#2. Reverse the the 2nd part of the list
#3. Append accordingly

def reorderList(self, head: Optional[ListNode]) -> None:
    """
    Do not return anything, modify head in-place instead.
    """
    
    #Using fast and slow pointers to locate the mid point
    slow, fast = head, head
    while fast.next and fast.next.next:
        slow = slow.next
        fast = fast.next.next
        
    #Reversing the 2nd part of the list   
    #Slow is at end of first part
    curr=slow.next
    #Classic reverse list from above
    prev=None
    while curr:
        temp = curr.next
        curr.next = prev
        prev = curr
        curr = temp
    
    #Making the last node of first part as None as this will be the end 
    slow.next = None
    #List 2
    head2 = prev
    #List 1
    head1 = head
    #Adjusting accordingly as per the given structure
    while head1:
        temp = head1.next
        head1.next = head2
        head1 = head2
        head2 = temp
```
